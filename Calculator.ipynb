##### Calculator - A simple calculator to do basic arithmetic operations. Make it a scientific calculator for added complexity.


```python
import tkinter
from tkinter import *
```

```python
def one():
    result.insert(tkinter.END, "1")
    
def two():
    result.insert(tkinter.END,"2")
    
def three():
    result.insert(tkinter.END,"3")
    
def four():
    result.insert(tkinter.END,"4")
    
def five():
    result.insert(tkinter.END,"5")
    
def six():
    result.insert(tkinter.END,"6")
    
def seven():
    result.insert(tkinter.END,"7")
    
def eight():
    result.insert(tkinter.END,"8")
    
def nine():
    result.insert(tkinter.END,"9")
    
def zero():
    result.insert(tkinter.END,"0")
    
def add():
    result.insert(tkinter.END, "+")

def subtract():
    result.insert(tkinter.END, "-")

def multiply():
    result.insert(tkinter.END, "*")

def divide():
    result.insert(tkinter.END, "/")
    
def point():
    result.insert(tkinter.END, ".")
```

```python
def clear_all():
    result.delete(0, tkinter.END)
```

```python
def backspace():
    current_text = result.get()
    updated_text = current_text[:-1]
    result.delete(0, tkinter.END)
    result.insert(tkinter.END, updated_text)
```

```python
def get_result():
    try:
        # Get the mathematical expression from the Entry widget
        expression = result.get()
        
        # Evaluate the expression using eval()
        result_value = eval(expression)
        
        # Clear the Entry widget and insert the result
        result.delete(0, tkinter.END)
        result.insert(tkinter.END, str(result_value))
    except Exception as e:
        # Handle any exceptions, for example, division by zero
        result.delete(0, tkinter.END)
        result.insert(tkinter.END, "Error")
```

```python
def percentage():
    current_value = result.get()
    try:
        # Convert the current value to a float and calculate the percentage
        current_value = float(current_value)
        percentage_value = current_value / 100.0

        # Update the entry widget with the calculated percentage
        result.delete(0, tkinter.END)
        result.insert(tkinter.END, str(percentage_value))
    except ValueError:
        # Handle the case where the current value is not a valid number
        result.delete(0, tkinter.END)
        result.insert(tkinter.END, "Error")

        
```

```python
window = Tk()
window.title('Calculator')
window.geometry('300x500')

button10 = Button(window,text ='C',bg = 'red',borderwidth = '15', command = clear_all)
button10.place(x=10,y=100)

button11 = Button(window,bg = 'orange', text ='%',borderwidth = '15', command = percentage)
button11.place(x=85,y=100)

button12 = Button(window,bg = 'orange', text ='\u2190',borderwidth = '15',command = backspace)
button12.place(x=160,y=100)

button13 = Button(window,bg = 'orange' , text ='/',borderwidth = '15', command = divide)
button13.place(x=235,y=100)

button7 = Button(window,bg = 'lightblue', text ='7',borderwidth = '15',command = seven)
button7.place(x=10,y=170)

button8 = Button(window,bg = 'lightblue',text ='8',borderwidth = '15',command = eight)
button8.place(x=85,y=170)

button9 = Button(window,bg = 'lightblue',text ='9',borderwidth = '15',command = nine)
button9.place(x=160,y=170)

button14 = Button(window,bg = 'orange' ,text ='x',borderwidth = '15', command = multiply)
button14.place(x=235,y=170)

button4 = Button(window,bg = 'lightblue',text ='4',borderwidth = '15',command = four)
button4.place(x=10,y=240)

button5 = Button(window,bg = 'lightblue',text ='5',borderwidth = '15',command =five)
button5.place(x=85,y=240)

button6 = Button(window,bg = 'lightblue',text ='6',borderwidth = '15',command = six)
button6.place(x=160,y=240)

button15 = Button(window,bg = 'orange', text ='-',borderwidth = '15', command = subtract)
button15.place(x=235,y=240)

button1 = Button(window,bg = 'lightblue',text ='1',borderwidth = '15',command = one)
button1.place(x=10,y=310)

button2 = Button(window,bg = 'lightblue',text ='2',borderwidth = '15',command = two)
button2.place(x=85,y=310)

button3 = Button(window,bg = 'lightblue',text ='3',borderwidth = '15', command = three)
button3.place(x=160,y=310)

button16 = Button(window,bg = 'orange',text ='+',borderwidth = '15', command = add)
button16.place(x=235,y=310)

button17 = Button(window,bg = 'lightblue',text ='0',borderwidth = '15', command = zero)
button17.place(x=45,y=380)

button18 = Button(window,bg = 'lightblue',text ='.',borderwidth = '15',command = point)
button18.place(x=130,y=380)

button19 = Button(window,bg = 'lightgreen',text ='=',borderwidth = '15',command = get_result)
button19.place(x=200,y=380)

result = Entry(window,borderwidth='10', bg = 'lavender')
result.place(x=60,y=10)





window.mainloop()
```
